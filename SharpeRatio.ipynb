```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

# List of Indian stock tickers
tickers = [
    'ADANIPORTS.NS', 'AMBUJACEM.NS', 'ASIANPAINT.NS', 'AURIONPRO.NS',
    'AXISBANK.NS', 'BAJAJ-AUTO.NS', 'BAJFINANCE.NS', 'BAJAJFINSV.NS', 'BHARTIARTL.NS',
    'BIOCON.NS', 'BPCL.NS', 'BRITANNIA.NS', 'CIPLA.NS', 'DIVISLAB.NS', 'DRREDDY.NS',
    'EICHERMOT.NS', 'GAIL.NS', 'GRASIM.NS', 'HCLTECH.NS', 'HDFCBANK.NS', 'HEROMOTOCO.NS',
    'HINDALCO.NS', 'HINDUNILVR.NS', 'ICICIBANK.NS',  'INDUSINDBK.NS',
    'INFY.NS', 'ITC.NS', 'JSWSTEEL.NS', 'KOTAKBANK.NS', 'M&M.NS', 'MARUTI.NS', 'NESTLEIND.NS',
    'NTPC.NS', 'ONGC.NS', 'POWERGRID.NS', 'RELIANCE.NS', 'SHREECEM.NS', 
    'SUNPHARMA.NS', 'TATAMOTORS.NS', 'TATACONSUM.NS', 'TITAN.NS', 'ULTRACEMCO.NS'
]

# Function to calculate Sharpe Ratio
def calculate_sharpe_ratio(returns, risk_free_rate=0):
    avg_return = returns.mean()
    std_dev = returns.std()
    if std_dev == 0:
        return np.nan
    sharpe_ratio = (avg_return - risk_free_rate) / std_dev
    return sharpe_ratio

# Dictionary to store Sharpe Ratios
sharpe_ratios = {}

# Download data, calculate rolling Sharpe Ratio, and store in dictionary
for ticker in tickers:
    data = yf.download(ticker, start='2013-01-01', end='2023-01-01')
    data['Daily Return'] = data['Adj Close'].pct_change()
    rolling_sharpe = data['Daily Return'].rolling(window=10).apply(calculate_sharpe_ratio, raw=False)
    sharpe_ratios[ticker] = rolling_sharpe.dropna()

# Convert dictionary to DataFrame
sharpe_ratios_df = pd.DataFrame(sharpe_ratios)

# Save to CSV
sharpe_ratios_df.to_csv('stocks_rolling_sharpe_ratios.csv')

# Plotting example for the first ticker
plt.figure(figsize=(14, 7))
plt.plot(sharpe_ratios[tickers[0]], label=f'10-Day Rolling Sharpe Ratio of {tickers[0]}', color='blue')
plt.title(f'10-Day Rolling Sharpe Ratio of {tickers[0]} from 2013 to 2023')
plt.xlabel('Date')
plt.ylabel('Sharpe Ratio')
plt.legend()
plt.grid(True)
plt.show()
```
